### Categorization according to the price.

According to price range analysis I would use these percentiles to divide items into price categories:

```python
print("25th percentile:", 
       np.percentile(products['unit_price'], 25))
print("50th percentile:", 
       np.percentile(products['unit_price'], 50))
print("75th percentile:", 
       np.percentile(products['unit_price'], 75))
print("95th percentile:", 
       np.percentile(products['unit_price'], 95))
print("99th percentile:", 
       np.percentile(products['unit_price'], 99))
```

```python
def price_cat(price):
    if price == 0:
        return 'zero'
    if 0 < price <= 1.25:
        return 'low'
    if 1.25 < price <= 2.08:
        return 'average low'
    if 2.08 < price <= 4.13:
        return 'average'
    if 4.13 < price <= 9.95:
        return 'average high'
    if 9.95 < price <= 16.63:
        return 'high'
    return 'very high'

products['price_category'] = products['unit_price'].apply(price_cat)
products['price_category'].value_counts()
```

According to price distribution and percentiles found above, I've divided all prices to these categories:

* zero price, 573 items ordered
* low (below 1.25), 160534 items ordered
* average low (between 1.25 and 2.08), 102337 items ordered
* average (between 2.08 and 4.13), 137107 items ordered
* average high (between 4.13 and 9.95), 99804 items ordered
* high (between 9.95 and 16.63), 17595 items ordered
* very high (above 16.63), 4743 items ordered    

Most items sold are in low price category.

### Check if there is a seasonal sales: analyze if there is a change in sales of the products according to the season and check it's sales amounts during summer and winter.

```python
Y = 2000 # dummy leap year to allow input X-02-29 (leap day)
seasons = [('winter', (date(Y,  1,  1),  date(Y,  2, 28))),
           ('spring', (date(Y,  3, 1),  date(Y,  5, 31))),
           ('summer', (date(Y,  6, 1),  date(Y,  8, 31))),
           ('autumn', (date(Y,  9, 1),  date(Y, 10, 31))),
           ('winter', (date(Y, 11, 1),  date(Y, 12, 31)))]

def get_season(now):
    if isinstance(now, datetime):
        now = now.date()
    now = now.replace(year=Y)
    return next(season for season, (start, end) in seasons
                if start <= now <= end)

print(get_season(date.today()))
```

```python
cohort_sizes = ltv_table.groupby('first_order_month').agg({'customer_id': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()
```

## Testing Hypothesis

### Average purchase size among top two product categories.

```python
# Creating groups for the test:
vintage_revenue=products[products['clear_category']=='vintage']['total_revenue']
kitchen_revenue=products[products['clear_category']=='kitchen']['total_revenue']

# Checking the length of each sample
print(len(vintage_revenue))
print(len(kitchen_revenue))
```

Let's see if the average purchase size within the groups differ.

* H0: the average purchase size is the same for both groups.
* H1: the average purchase size is different for both groups. 

Let's perform the test with 5% significance level.

```python
stat, p = mannwhitneyu(vintage_revenue, kitchen_revenue)

print('Statistics=%.3f, p=%.5f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
```

We fail to reject the null hypothesis as the p-value is greater than the alpha. I can say that kitchen category has equal average purchase size per item as compared to vintage.

### Average purchase price per unit among top two product categories.

```python
# Creating groups for the test:
kitchen_perprice=products[products['clear_category']=='kitchen']['unit_price']
vintage_perprice=products[products['clear_category']=='vintage']['unit_price']

# Checking the length of each sample
print(len(kitchen_perprice))
print(len(vintage_perprice))
```

Let's see if the average purchase price per unit within the groups differ.

* H0: the average purchase price per unit is same for both groups.
* H1: the average purchase price per unit is different for both groups. 

Let's perform the test with 5% significance level.

```python
stat, p = mannwhitneyu(kitchen_perprice, vintage_perprice)

print('Statistics=%.3f, p=%.5f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
```

We reject the null hypothesis as the p-value is less than the alpha. I can say that kitchen category has different average purchase price per unit as compared to vintage category.
